```python
%sql
select
    p.product_id,
    p.product_name,
    p.brand_id,
    -- p.category_id,
    -- c.category_id as c_category_id,
    c.category_name,
    p.model_year,
    p.list_price,
    p.bronze_ingest_timestamp,
    p.bronze_source_file    
from tmp_bronze_products as p
LEFT JOIN tmp_bronze_categories as c
ON p.category_id = c.category_id
```

```python
df_product_silver = spark.sql(""" 
  select
      p.product_id,
      p.product_name,
      c.category_name,
      p.brand_id,
      -- p.category_id,
      -- c.category_id as c_category_id,
      p.model_year,
      p.list_price,
      p.bronze_ingest_timestamp,
      p.bronze_source_file    
  from tmp_bronze_products as p
    LEFT JOIN tmp_bronze_categories as c
      ON p.category_id = c.category_id    
""")
```

```python
display(df_product_silver)
```

```python
#salvar em parquet como delta  tabela de teste SILVER
df_product_silver.write\
    .mode("overwrite")\
    .format("delta")\
    .option("mergeSchema", "true")\
    .saveAsTable(f"{silver_path}.product_teste")

```
