```python
# Definir pastas do projetos em variaveis para facilitar
bronze_path   = 'bikestore.bronze'
silver_path   = 'bikestore.silver'
gold_path     = 'bikestore.gold'
resource_path = "/Volumes/bikestore/resource/origem"
resource_path_volume = '/Volumes/bikestore/resource/origem'
```

```python
from pyspark.sql.functions import current_timestamp, col

df_stores = spark.read.csv( 
                            f"{resource_path}/stores.csv",
                            header = True, inferSchema = True)\
                            .withColumn("bronze_ingest_timestamp", current_timestamp())\
                            .withColumn("bronze_source_file", col("_metadata.file_path"))
```

```python
#salvar em parquet como delta 
df_stores.write\
    .mode('overwrite')\
    .format('delta')\
    .option("mergeSchema", "true") \
    .saveAsTable(f"{bronze_path}.stores")
```
